```python
root_dir = r'C:\Users\ys\Documents\WeChat Files\wxid_1gx7h791samv22\FileStorage\File\2021-10\VIS.csv'
import pandas as pd 
import numpy as np 
import os
import re
```

```python
data = pd.read_csv(root_dir)
data.drop(labels=0,axis=0,inplace= True)
columns = data.columns
data
```

```python
data = np.array(data,dtype=np.float32)
df = pd.DataFrame()
for i in range(10):
    wavelength = data[:,4*i]
    y_0 = data[:,4 * i+1]
    y_0 = y_0 - y_0[0]
    y_1 = data[:,4*i+3]
    y_1 = y_1 - y_1[0]
    out = y_1 -y_0
  
    if i ==0:
        df.insert(value = wavelength,loc=i,column = "Wavelength")
    df.insert(loc=i+1,value=out,column= columns[4*i])

df.to_csv(r'C:\Users\ys\Desktop\VIS_out.csv')

    
```
