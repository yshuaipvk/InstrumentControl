```python
import numpy as np 
import pandas as pd 
import re
import os
import matplotlib.pyplot as plt 
root_dir = "D:/ys/DATA/20211003-CPTA-OCVD-TEM"
fileNames = os.listdir(root_dir)
fileNames
```

```python
from trce_Ion import get_voltage_recovery
tem_ion_files0 = []
vrs = []
for i in range(len(fileNames)):
    filename = fileNames[i]
    if re.findall('K-',filename) and re.findall('ion',filename) and re.findall('-0-',filename):
        tem_ion_files0.append(filename)


df = pd.DataFrame()
i = 0
for filename in tem_ion_files0:
    tem = filename[:4]
    file_dir = os.path.join(root_dir,filename)
    data = np.loadtxt(file_dir,delimiter=',')
    Vr = get_voltage_recovery(data)
    x = data[:,0]
    y = data[:,1]
    if i ==0:
        df.insert(loc=i,value=x,column = 'time')
    df.insert(loc=i+1,value=y,column=tem)
    i+=1
out_dir = os.path.join(root_dir,'tem_ion.csv')
df.to_csv(out_dir,index=None)

        
```
